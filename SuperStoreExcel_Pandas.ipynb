```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_excel("C:/Users/Gabriel Johnson/Downloads/SuperStoreData-230102-093001.xlsx")
data.head()
```

```python
# Categorical Profit in Los Angeles

la = data[(data['City'] == 'Los Angeles')]

la
```

```python
# Grouping under Categories
la_category = la.groupby(by='Category').sum()
la_category
```

```python
la_category.Profit
```

```python
label = la_category.Profit.index
value = la_category.Profit
```

```python
# Chart Visualization
plt.pie(value, labels=label, colors=['blue', 'red', 'yellow'], autopct='%1.1f%%')
plt.title("Los Angeles Categorical Profit")
plt.show()
```

```python
# Top 10 Most Profitable Customers
# Grouping Data
data.groupby(by='Customer Name').sum()
```

```python
customers = data.groupby(by='Customer Name').sum()
# Sorting in Descending Order
customers.sort_values(by='Profit', ascending = False)
```

```python
# Top 10
customers.sort_values(by='Profit', ascending = False).head(10)
```

```python
# Bar Chart Visualization
top_10 = customers.sort_values(by='Profit', ascending = False).head(10)
top_10.Profit # top_10['Profit ID'] (If there's Space) 
```

```python
# Values & Labels
value = top_10.Profit
label = top_10.Profit.index

label
```

```python
# Plotting Bar Chart (NB: Pie Chart 'Value before Label', Bar Chart 'Label before Value')
plt.figure(figsize=(7, 3))
plt.xticks(rotation=45)
plt.bar(label, value, color='orange')
plt.xlabel('Customer Name')
plt.ylabel('Profit')
plt.title("Top 10 Most Profitable Customers")
plt.show()
```
